```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
df=pd.read_excel(r'C:\Users\dhivy\Desktop\machine learning\stu\Task____students_performance_dataset (2).xlsx')
print(df.shape)
print(df.head())
```

```python
print("Column Names",df.columns.tolist())
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df=df.drop('Student_ID',axis=1)
```

```python

print(df.shape)
```

```python

print(df.dtypes)
```

```python
cat_data=['Gender','Parental_Education','Internet_Access','Tutoring_Classes','Sports_Activity','Extra_Curricular','School_Type','Teacher_Feedback']
num_data=['Study_Hours_per_Week','Attendance_Percentage','Previous_Sem_Score','Family_Income','Sleep_Hours','Travel_Time','Test_Anxiety_Level','Peer_Influence','Motivation_Level']
```

```python
le=LabelEncoder()
for col in cat_data:
    df[col]=le.fit_transform(df[col])
print("Categorical Data is Encoded")
```

```python
scalar=StandardScaler()
df[num_data]=scalar.fit_transform(df[num_data])
print("Numerical are Scaled")
df.head()
```

```python
x=df.drop('Final_Score',axis=1)
y=df['Final_Score']
print(x.shape)

print(y.shape)
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,random_state=42,test_size=0.2)
print(f"Testing Data: {x_test.shape[0]}")
print(f"Training Data: {x_train.shape[0]}")
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
print("Model Trainied Successfully")
```

```python
y_train_pred=model.predict(x_train)
y_test_pred=model.predict(x_test)

print("Performance metrics")
print("Training Data")
print(f"MAE:{mean_absolute_error(y_train,y_train_pred):.4f}")
print(f"MSE:{mean_squared_error(y_train,y_train_pred):.4f}")
print(f"R2:{r2_score(y_train,y_train_pred):.4f}")

print("Testing Data")
print(f"MAE:{mean_absolute_error(y_test,y_test_pred):.4f}")
print(f"MSE:{mean_squared_error(y_test,y_test_pred):.4f}")
print(f"R2:{r2_score(y_test,y_test_pred):.4f}")
```

```python
print("Sample Predictions:")
print("-" * 50)
sample_indices = [0, 50, 100, 150, 200]
for idx in sample_indices:
    if idx < len(y_test):
        pred = y_test_pred[idx]
        actual = y_test.iloc[idx]
        print(f"Actual: {actual:.2f} | Predicted: {pred:.2f} | Error: {abs(actual-pred):.2f}")
```

```python
numeric_cols_for_scaling = [col for col in num_data if col != 'Final_Score']
joblib.dump(model, "student_model.pkl")
joblib.dump(scalar, "student_scaler.pkl")
joblib.dump(x.columns, "student_feature_order.pkl")
joblib.dump(numeric_cols_for_scaling, "student_numeric_columns.pkl")
print("Model and preprocessing objects saved successfully!")
```
